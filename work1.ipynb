```python
# df['International Reputation'].mean()
df['International Reputation'].fillna(1, inplace = True)
# df['Skill Moves'].median()
df['Skill Moves'].fillna(df['Skill Moves'].median(), inplace = True)
df['Weak Foot'].fillna(round(df['Weak Foot'].mean()), inplace = True)
df['Crossing'].fillna(round(df['Crossing'].mean()), inplace = True)
df['Finishing'].fillna(round(df['Finishing'].mean()), inplace = True)
df['HeadingAccuracy'].fillna(round(df['HeadingAccuracy'].mean()), inplace = True)
df['ShortPassing'].fillna(round(df['ShortPassing'].mean()), inplace = True)
df['Volleys'].fillna(round(df['Volleys'].mean()), inplace = True)
df['Dribbling'].fillna(round(df['Dribbling'].mean()), inplace = True)
df['Curve'].fillna(round(df['Curve'].mean()), inplace = True)
df['FKAccuracy'].fillna(round(df['FKAccuracy'].mean()), inplace = True)
df['LongPassing'].fillna(round(df['LongPassing'].mean()), inplace = True)
df['BallControl'].fillna(round(df['BallControl'].mean()), inplace = True)
df['Acceleration'].fillna(round(df['Acceleration'].mean()), inplace = True)
df['SprintSpeed'].fillna(round(df['SprintSpeed'].mean()), inplace = True)
df['Agility'].fillna(round(df['Agility'].mean()), inplace = True)
df['Reactions'].fillna(round(df['Reactions'].mean()), inplace = True)
df['Balance'].fillna(round(df['Balance'].mean()), inplace = True)
df['ShotPower'].fillna(round(df['ShotPower'].mean()), inplace = True)
df['Jumping'].fillna(round(df['Jumping'].mean()), inplace = True)
df['Stamina'].fillna(round(df['Stamina'].mean()), inplace = True)
df['Strength'].fillna(round(df['Strength'].mean()), inplace = True)
df['LongShots'].fillna(round(df['LongShots'].mean()), inplace = True)
df['Aggression'].fillna(round(df['Aggression'].mean()), inplace = True)
df['Interceptions'].fillna(round(df['Interceptions'].mean()), inplace = True)
df['Positioning'].fillna(round(df['Positioning'].mean()), inplace = True)
df['Vision'].fillna(round(df['Vision'].mean()), inplace = True)
df['Penalties'].fillna(round(df['Penalties'].mean()), inplace = True)
df['Composure'].fillna(round(df['Composure'].mean()), inplace = True)
df['Marking'].fillna(round(df['Marking'].mean()), inplace = True)
df['StandingTackle'].fillna(round(df['StandingTackle'].mean()), inplace = True)
df['SlidingTackle'].fillna(round(df['SlidingTackle'].mean()), inplace = True)
```

```python
unnamed_columns = df.columns[df.columns.str.contains('^Unnamed')]
df = df.drop(unnamed_columns, axis=1)
df.shape
```

```python
def value_and_wage_conversion(Value):
    if isinstance(Value,str):
        out = Value.replace('â‚¬', '')
        if 'M' in out:
            out = float(out.replace('M', ''))*1000000
        elif 'K' in Value:
            out = float(out.replace('K', ''))*1000
        return float(out)


df['Value'] = df['Value'].apply(lambda x: value_and_wage_conversion(x))
df['Wage'] = df['Wage'].apply(lambda x: value_and_wage_conversion(x))
df['Release Clause'] = df['Release Clause'].apply(lambda x: value_and_wage_conversion(x))
df['Release Clause'].fillna(df['Release Clause'].mean(), inplace = True)
```

```python
df['Club'].fillna('No Club', inplace = True)
```

```python
def clean_date(x):
    l=[]
    if isinstance(x,str):
        l=x.split(", ")
        x=l[-1]
    return(x)

df['Joined']=df['Joined'].replace(np.nan,0)
df['Joined'] = df['Joined'].apply(clean_date).astype('str')
```

```python
def clean_weight(x):

    if isinstance(x,str):
        return(x.replace('lbs', ''))
    return(x)

df['Weight'] = df['Weight'].apply(clean_weight).astype('float')
df['Weight'].fillna(round(df['Weight'].mean()), inplace = True)
```

```python
df['Work Rate'].value_counts()
df['Work Rate'].fillna('Medium/ Medium', inplace = True)
```

```python
df['Preferred Foot'].value_counts()
df['Preferred Foot'].fillna('Right', inplace = True)
```

```python

df['Body Type'].fillna('Normal', inplace = True)
df['Body Type'].value_counts()
df.info()
```

```python
def skillConverter(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    else:
        return val

skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']

for col in skill_columns:
    df[col] = df[col].apply(skillConverter)
    df[col].fillna(0.0, inplace = True)
```

```python
df['Contract Valid Until'].fillna('Not available', inplace = True)
df.info()
```

```python
fifa=df
def clean_height(x):
    l=[]
    s=1.0
    if isinstance(x,str):
        l=x.split("'")
        i=int(l[0])
        j=int(l[1])
        s=((i*12)+j)*2.54
    return(s)


fifa['Height'] = fifa['Height'].apply(clean_height)
fifa['Height'].fillna(round(fifa['Height'].mean()), inplace = True)
```
